```
class PriorityQueue:
	def __init__(self):
		self.__data = [[]]
		self.__size = 0
	def is_empty(self):
		return len(self.__data) == 0
	def length(self):
		return len(self.__data)
	def enqueue(self,prioridad , elem):
		'''agregar'''
		if (type(prioridad) is not int) or (prioridad < 0) :
			print('Fuera de rango')
		else:
			if prioridad > len(self.__data):
				for i in range(prioridad - len(self.__data)+3):
					self.__data.append([])
			self.__data[prioridad].append(elem)
			self.__size += 1

	def dequeue(self):
		'''sacar'''
		for i in self.__data:
			if len(i) > 0:
				self.__size -= 1
				return i.pop(0)

	def to_string(self):
		cont= 0
		for i in self.__data:
			for j in i:
				if cont == 0:
					print(f'|{j}| <--- head')
				elif cont == self.__size-1:
					print(f'|{j}| <--- tail')
				else:
					print(f'|{j}|')
				cont+=1

		print(' ')

barco = PriorityQueue()
barco.enqueue(4,'Maestre')
barco.enqueue(2,'Niños')
barco.enqueue(4,'Mecánico')
barco.enqueue(3,'Mujeres')
barco.enqueue(2,'3ra edad')
barco.enqueue(1,'Niñas')
barco.enqueue(3,'Hombres')
barco.enqueue(4,'Vigia')
barco.enqueue(5,'Capitán')
barco.enqueue(4,'Timonel')
barco.enqueue(1,'animales')
barco.enqueue(0,'bebés')
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
barco.dequeue()
barco.to_string()
print("No hay nadie en la cola")
print("El barco ya ha sido evacuado por completo")
```
